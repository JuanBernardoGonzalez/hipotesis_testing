## Elijo variables y hago las pruebas de hipótesis

#### Una variable

- Parace lógico que variables como W (victorias), Pts (puntos) o Pts/G (puntos/partidos jugados) estén sesgadas y por tanto no se distribuyan como normal. Por ello, una variable como GD(diferencia entre goles anotados y recibidos), me interesa. GD presenta en la descripción una media y una mediana parecidas. Al revisar sus gráficos, GD presenta una distribución que parece simétrica, por lo que haré una prueba de normalidad.  

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))
sns.scatterplot(UEFA['GD'], ax=axes[1, 0],color= 'black')
sns.boxplot(UEFA['GD'], ax=axes[1, 1])
sns.histplot(UEFA['GD'], ax=axes[0, 0])
sns.kdeplot(UEFA['GD'], color= 'black', fill = True, ax=axes[0, 1])
plt.show()
```

- La prueba de normalidad será la de Shapiro-Wilk. Esta prueba asume que las observaciones en cada muestra son independientes y están distribuidaas idénticamente. Al ser observaciones de una variable en una sola muestra, podemos decir que las observaciones cumplen con las asunciones de la prueba.  

```python
from scipy.stats import shapiro
data = UEFA['GD']
stat, p = shapiro(data)
#print('stat=%.3f, p=%.3f' % (stat, p))
print('p=%.3f' % p)
if p > 0.05:
 print('Es probable que se distribuya como normal')
else:
 print('Es probable que no se distribuya como normal')
```

- Al ser el valor p mayor que un alfa de 0.05 podemos decir que con un 95% de confianza las observaciones de GD se distribuyen como normal. 

# Ejercicio 2: dos variables

```python
C = UEFA[['W','D','L','GF','GA','GD','xGA','xGD','xGD/90']]
sns.heatmap(C.corr(),annot = True, annot_kws = {'size': 6})
plt.show()
```

- Al revisar las correlaciones en el mapa de colores, la correlación más baja es de 0.07 para las variables D (empates) y L (perdidos). Voy a a hacer la prueba de Sperman (relación monotónica) para verificar si D Y L son o no independientes. 
- La prueba de Sperman asume que las observaciones en cada muestra son independientes y que puden ser clasificadas (ranked). Desde que ambas variables tienen una correlación cercana a cero, sus observaciones vienen de una muestra y pertenecen a equipos diferentes y partidos diferentes, podemos asumir que son independientes. Al ser números que nos dan la cantidad de partidos perdidos o empatados en un periodo de tiempo, son observaciones que se pueden clasificar (ranked). 

```python
from scipy.stats import spearmanr
stat, p = spearmanr(UEFA['D'], UEFA['L'])
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
 print('Probablemente independientes')
else:
 print('Probablemente dependientes')
```

- Con una p mayor a alfa de 0.05 podemos concluir que D y L no tienen una relación monotónica. Con un 95% de probabilidad D y L son independientes. 

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))
sns.kdeplot(UEFA['D'], ax=axes[0, 0], color= 'b', fill = True, linewidth=2)
sns.kdeplot(UEFA['L'],  ax=axes[0, 1], color= 'g', fill = True, linewidth=2)
sns.kdeplot(UEFA['D'], bw_method=0.5, fill = True, ax=axes[1, 0])
sns.kdeplot(UEFA['L'], bw_method=0.5, fill = True, ax=axes[1, 0])
sns.histplot(UEFA['D'], ax=axes[1, 1], color= 'b')
sns.histplot(UEFA['L'], ax=axes[1, 1], color= 'y')
plt.show()
```

# Ejercicio 3: tres variables

- Después de conluir la independencia de L y D, podemos investigar sus relaciones con GA(goles en contra). GA tiene una correlación casi cero con D y está positiva y fuertemente correlacionado con L -los equipos que más pierden son los que más goles reciben-. Veamos si L, D y GA vienen de distribuciones distintas y si sus medias son diferentes. Voy a hacer el Test ANOVA.


```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))
sns.kdeplot(UEFA['D'], ax=axes[0, 0], color= 'b', fill = True, linewidth=2)
sns.kdeplot(UEFA['L'],  ax=axes[0, 1], color= 'g', fill = True, linewidth=2)
sns.kdeplot(UEFA['GA'],  ax=axes[1, 0], color= 'y', fill = True, linewidth=2)
sns.kdeplot(UEFA['D'], bw_method=0.5, color='b', fill = True, ax=axes[1, 1])
sns.kdeplot(UEFA['L'], bw_method=0.5, color='g', fill = True, ax=axes[1, 1])
sns.kdeplot(UEFA['GA'], bw_method=0.5, color='y', fill = True, ax=axes[1, 1])
plt.show()
```

- En ANOVA las observaciones en cada muestra han de ser independientes y distribuidas identicamente, han de distribuirse como normal y tener la misma varianza. 
- Las tres distribuciones parecen normales y la varianza de L es similar a la de D. Las varianzas de L y D parecen distintas a las de GA y sabemos que la correlación entre GA y L es alta. 
  - El supuesto de varianzas iguales entre grupos seguramente no se se cumple, pero el ANOVA, al tener un grupo relativamente grande (98 observaciones) con un diseño equilibrado (mismo número de observaciones por grupo), se hace bastante robusto a la falta de homocedasticidad. Para este caso podemos no cumplir el supuesto de varianzas iguales. 
  - GA es independiente a D pero no a L. El supuesto de independencia no se cumple. 
- Aún así ejecutaré el test (no encuentro tres variables que se distribuyan como normal y sean independientes entre ellas) 

```python
from scipy.stats import f_oneway
stat, p = f_oneway(UEFA['D'], UEFA['L'], UEFA['GA'])
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
 print('Probablemente la misma distribución Ho:las medias de las muestras son iguales')
else:
 print('Probablemente diferentes distribuciones H1:las medias de una o más muestras son desiguales')
```

- Se acepta la hipótesis alternativa H1 con alfa de 0.05. Con un 95% de probabilidad una o más muestras son desiguales.
